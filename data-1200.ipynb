```python
quality = pd.read_csv('./quality.csv')
```

```python
quality.head
```

```python
merged_df = filtered_df.merge(quality[['배정번호', '양품수량','불량수량','총수량']], on='배정번호', how='inner')
```

```python
merged_df.head
```

```python
merged_df['불량율'] = merged_df['불량수량'] / merged_df['총수량']
```

```python
merged_df.head
```

```python
# merged_df로 그룹화하고 불량율의 평균을 계산
average_defect_rate_per_assignment = merged_df.groupby('배정번호')['불량율'].mean()
print(average_defect_rate_per_assignment)
```

```python
q3 = average_defect_rate_per_assignment.quantile(0.75)
```

```python
merged_df['상태'] = '안정'  # 모든 행을 '안정'으로 초기화
merged_df.loc[merged_df['불량율'] > q3, '상태'] = '위험'  # q3보다 큰 경우 '위험'으로 설정
```

```python
stable_assignments = merged_df[merged_df['상태'] == '안정']['배정번호'].unique()
len(stable_assignments)
```

```python
unstable_assignments = merged_df[merged_df['상태'] == '위험']['배정번호'].unique()
len(unstable_assignments)
```
